```python
import numpy as np

def drag_coefficient(angle_of_attack):
    if 0 <= angle_of_attack <= 11:
        return 0.11
    elif 11 < angle_of_attack <= 90:
        return (1.35 / 6400) * (angle_of_attack - 10) ** 2 + 0.15
    elif 90 < angle_of_attack <= 170:
        return (1.35 / 6400) * (angle_of_attack - 170) ** 2 + 0.15
    else:
        return 0

def lift_coefficient(angle_of_attack):
    if 0 < angle_of_attack <= 11:
        return 0  # Assuming zero lift at low angles based on JavaScript function
    elif 11 < angle_of_attack <= 32:
        return -(1 / 300) * (angle_of_attack - 32) ** 2 + 1.5
    elif 32 < angle_of_attack <= 90:
        return -(1.3 / 3364) * (angle_of_attack - 32) ** 2 + 1.5
    else:
        return 0

import matplotlib.pyplot as plt

# Create a range of angle of attack values
angles = np.linspace(0, 180, 1000)
# Calculate the drag coefficient for each angle
drag_values = [drag_coefficient(angle) for angle in angles]
lift_values = [lift_coefficient(angle) for angle in angles]

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(angles, drag_values, 'b-', linewidth=2)
# plt.plot(angles, lift_values, 'b-', linewidth=2)
plt.title('Drag Coefficient vs Angle of Attack')
plt.xlabel('Angle of Attack (degrees)')
plt.ylabel('Drag Coefficient')
plt.grid(True)
plt.axhline(y=0, color='k', linestyle='-', alpha=0.3)
plt.axvline(x=0, color='k', linestyle='-', alpha=0.3)
plt.ylim(0, 1.6)
plt.show()
```
